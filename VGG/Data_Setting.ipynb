```python
import os

folder_path = '/Users/stepground/Downloads/New_sample/원천데이터/DOG/BODYLOWER'  # 검색할 폴더의 경로를 지정합니다.

# 폴더 내의 파일 및 폴더 목록을 가져옵니다.
contents = os.listdir(folder_path)

# 파일과 폴더를 구분하여 출력합니다.
for content in contents:
    full_path = os.path.join(folder_path, content)  # 항목의 전체 경로를 얻습니다.
    if os.path.isfile(full_path):
        print(f'파일: {content}')
    elif os.path.isdir(full_path):
        print(f'폴더: {content}')
    else:
        print(f'기타: {content}')
```

```python
import os
import shutil

# 원본 폴더 경로 설정
source_folder = '/Users/stepground/Downloads/New_sample/원천데이터/DOG/'

# 대상 폴더 경로 설정
destination_folder = '/Users/stepground/Inha-Project/VGG/Images'

# 하위폴더 내의 이미지 파일을 복사하는 함수 정의
def copy_images_from_subfolders(src, dst, category):
    for root, dirs, files in os.walk(src):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif')):
                source_path = os.path.join(root, file)
                new_filename = f"{category}_{str(len(os.listdir(dst)) + 1)}{os.path.splitext(file)[1]}"
                destination_path = os.path.join(dst, category, new_filename)
                
                # 대상 폴더(category)가 없으면 생성
                os.makedirs(os.path.join(dst, category), exist_ok=True)
                
                # 중복 파일 이름 확인 및 다른 이름으로 저장
                if os.path.exists(destination_path):
                    base, ext = os.path.splitext(new_filename)
                    i = 1
                    while os.path.exists(destination_path):
                        new_filename = f"{base}_{i}{ext}"
                        destination_path = os.path.join(dst, category, new_filename)
                        i += 1
                
                shutil.copy(source_path, destination_path)
                print(f'이미지 복사: {file} -> {new_filename}')

# 'BODYLOWER' 폴더 내의 이미지 복사
bodylower_folder = os.path.join(source_folder, 'BODYLOWER')
for subdir in os.listdir(bodylower_folder):
    subdir_path = os.path.join(bodylower_folder, subdir)
    if os.path.isdir(subdir_path):
        copy_images_from_subfolders(subdir_path, destination_folder, 'bodylower')

# 'SIT' 폴더 내의 이미지 복사
sit_folder = os.path.join(source_folder, 'SIT')
for subdir in os.listdir(sit_folder):
    subdir_path = os.path.join(sit_folder, subdir)
    if os.path.isdir(subdir_path):
        copy_images_from_subfolders(subdir_path, destination_folder, 'sit')

```

```python
folder_path = '/Users/stepground/Inha-Project/VGG/Images'  # 검색할 폴더의 경로를 지정합니다.

# 폴더 내의 파일 및 폴더 목록을 가져옵니다.
contents = os.listdir(folder_path)

# 파일과 폴더를 구분하여 출력합니다.
for content in contents:
    full_path = os.path.join(folder_path, content)  # 항목의 전체 경로를 얻습니다.
    if os.path.isfile(full_path):
        print(f'파일: {content}')
    elif os.path.isdir(full_path):
        print(f'폴더: {content}')
    else:
        print(f'기타: {content}')
```

```python
import os
import tensorflow as tf

# 이미지 파일이 있는 디렉토리 경로
directory_path = "/Users/stepground/Inha-Project/VGG/Images"

# 이미지 파일 목록을 가져옵니다.
file_list = os.listdir(directory_path)

# 레이블을 저장할 리스트
labels = []

# 파일 이름을 분석하여 레이블을 생성합니다.
for filename in file_list:
    if filename.startswith("BODYLOWER"):
        labels.append(0)
    elif filename.startswith("SIT"):
        labels.append(1)
    else:
        # 다른 파일은 어떤 레이블을 부여할지에 대한 규칙을 추가할 수 있습니다.
        labels.append(-1)  # 예를 들어 다른 파일의 경우 -1을 사용

# 레이블 리스트를 TensorFlow Tensor로 변환합니다.
labels = tf.constant(labels, dtype=tf.int64)

print("레이블 값:", labels)
```

```python
import pickle

# 피클 파일 경로
pickle_file_path = "/Users/stepground/Inha-Project/VGG/Labels/labels.pkl"

# 레이블 리스트를 피클 파일로 저장
with open(pickle_file_path, 'wb') as file:
    pickle.dump(labels, file)

print("레이블이 피클 파일로 저장되었습니다.")
```
